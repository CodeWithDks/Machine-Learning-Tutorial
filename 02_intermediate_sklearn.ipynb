#Intermediate Machine Learning With Sklearn


```python
# libraries import
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# Load dataset
df = pd.read_csv("/content/drive/MyDrive/DataSets/Chapter 1/tips.csv")
```

```python
df.tail()
```

```python
# encode smoker column
df['smoker_num'] = df['smoker'].map({'Yes':1,'No':0})
```

```python
df.head()
```

```python
# select the features
x = df[['total_bill','size','smoker_num']]
# select target
y = df['tip']
```

```python
# model load/call
model = LinearRegression()
```

```python
# train the model
model.fit(x,y)
```

```python
# lets test the model
model.predict([[34.23, 4,0]])
```

# Classification


```python
from sklearn.linear_model import LogisticRegression
```

```python
# feature selection
x = df[['total_bill','size','tip']]
# target
y = df['smoker']
```

```python
# model load/call
model = LogisticRegression()
# train the model
model.fit(x,y)
```

```python
# test the model
model.predict([[23,3,1]])
```
