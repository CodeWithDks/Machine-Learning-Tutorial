#Scikit-learn library basic usage


```python
# libraries import
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# simple linear regression
from sklearn.linear_model import LinearRegression
```

```python
# load the dataset
df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
# create a regression line plot using seaborn
sns.lmplot(x='tip', y='total_bill', data=df)
```

```python
# select the feature column from df
x = df[['tip']]
# select the target column from df
y = df['total_bill']
```

```python
# load/call the linear regression model
model = LinearRegression()
# fit the model means train the model
model.fit(x,y)
```

```python
# lets predict the tip for an unkown total_bill
model.predict([[3.3]])
```
